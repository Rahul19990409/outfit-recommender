```python
#Program to implement Non-linear Spatial Filtering using Built-in and
#user defined functions.
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Display function
def show_image(title, img, cmap_type='gray'):
   plt.figure(figsize=(6, 6))
   plt.title(title)
   plt.imshow(img, cmap=cmap_type)
   plt.axis('off')
   plt.show()

# 1. Built-in Median Filter using OpenCV
def median_filter_builtin(img, ksize=3):
   return cv2.medianBlur(img, ksize)

# 2. User-defined Median Filter
def median_filter_custom(img, ksize=3):
   padded_img = np.pad(img, pad_width=ksize//2, mode='edge')
   output = np.zeros_like(img)
   for i in range(img.shape[0]):
       for j in range(img.shape[1]):
           neighborhood = padded_img[i:i+ksize, j:j+ksize]
           output[i, j] = np.median(neighborhood)
   return output

# 3. User-defined Max Filter
def max_filter(img, ksize=3):
   padded_img = np.pad(img, pad_width=ksize//2, mode='edge')
   output = np.zeros_like(img)
   for i in range(img.shape[0]):
       for j in range(img.shape[1]):
           neighborhood = padded_img[i:i+ksize, j:j+ksize]
           output[i, j] = np.max(neighborhood)
   return output

# 4. User-defined Min Filter
def min_filter(img, ksize=3):
   padded_img = np.pad(img, pad_width=ksize//2, mode='edge')
   output = np.zeros_like(img)
   for i in range(img.shape[0]):
       for j in range(img.shape[1]):
           neighborhood = padded_img[i:i+ksize, j:j+ksize]
           output[i, j] = np.min(neighborhood)
   return output

# ---- Main Program ---
if __name__ == "__main__":
# Load grayscale image
   img_path = 'sunrise beach.jpg'  # Replace with actual path
   img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

   if img is None:
       print(f"[ERROR] Image not found at path: {img_path}")
   else:
       show_image("Original Image", img)
       # Apply filters
       median_cv = median_filter_builtin(img, 3)
       median_custom = median_filter_custom(img, 3)
       max_filtered = max_filter(img, 3)
       min_filtered = min_filter(img, 3)

       # Show results
       show_image("Median Filter (Built-in)", median_cv)
       show_image("Median Filter (Custom)", median_custom)
       show_image("Max Filter", max_filtered)
       show_image("Min Filter", min_filtered)
```
